## Extraccion de datos de Coingecko

```python
import requests
import pandas as pd

def obtener_proyectos_ai(max_paginas=10):
    categoria_id = "artificial-intelligence"
    proyectos = []
    for page in range(1, max_paginas + 1):
        url = "https://api.coingecko.com/api/v3/coins/markets"
        params = {
            "vs_currency": "usd",
            "category": categoria_id,
            "order": "market_cap_desc",
            "per_page": 50,
            "page": page
        }
        response = requests.get(url, params=params)
        if response.status_code != 200:
            print("Error al obtener datos de AI")
            break
        data = response.json()
        if not data:
            break
        for proyecto in data:
            proyecto["clasificacion"] = "ai"
        proyectos.extend(data)

    df_ai = pd.DataFrame(proyectos)
    return df_ai

df_ai = obtener_proyectos_ai()
print(f"Proyectos AI: {df_ai.shape[0]}")
```

## ExtracciÃ³n de Exchanges

```python
import requests
import pandas as pd
import time

def obtener_lista_exchanges():
    url = "https://api.coingecko.com/api/v3/exchanges"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print("Error al obtener la lista de exchanges")
        return []

def obtener_monedas_por_exchange(exchange_id):
    url = f"https://api.coingecko.com/api/v3/exchanges/{exchange_id}"
    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            data = response.json()
            tickers = data.get('tickers', [])
            monedas = set()
            for ticker in tickers:
                monedas.add(ticker['base'].lower())
            return list(monedas)
        else:
            print(f"[{exchange_id}] Error {response.status_code}")
            return []
    except Exception as e:
        print(f"[{exchange_id}] ExcepciÃ³n: {e}")
        return []

exchanges = obtener_lista_exchanges()

resultado = []
limite = 50

for i, exchange in enumerate(exchanges[:limite]):
    exchange_id = exchange['id']
    exchange_name = exchange['name']
    print(f"[{i+1}] {exchange_name} ({exchange_id})")

    monedas = obtener_monedas_por_exchange(exchange_id)
    if monedas:
        for moneda in monedas:
            resultado.append({
                "exchange": exchange_name,
                "exchange_id": exchange_id,
                "moneda": moneda
            })

    time.sleep(30)

df_exchanges_monedas = pd.DataFrame(resultado)
print(df_exchanges_monedas.head())
print(f"Total pares exchange-moneda: {df_exchanges_monedas.shape[0]}")
```

```python
def obtener_exchanges_por_moneda(coin_id):
    url = f"https://api.coingecko.com/api/v3/coins/{coin_id}/tickers"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return list({ticker['market']['name'] for ticker in data['tickers']})
    else:
        return []

exchanges_btc = obtener_exchanges_por_moneda("bitcoin")
print("Bitcoin se comercia en:", exchanges_btc)
```

## Visualizar la red creada

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

for _, row in df_completo.iterrows():
    G.add_node(row["id"], label=row["name"], clasificacion=row["clasificacion"],
               market_cap=row["market_cap"], volume=row["total_volume"])

for i, row_i in df_completo.iterrows():
    for j, row_j in df_completo.iterrows():
        if i >= j:
            continue
        if row_i["clasificacion"] == row_j["clasificacion"]:
            if (row_i["total_volume"] + row_j["total_volume"]) == 0:
                weight = 0
            else:
                weight = (row_i["market_cap"] + row_j["market_cap"]) / (row_i["total_volume"] + row_j["total_volume"])
            G.add_edge(row_i["id"], row_j["id"], tipo="categoria", weight=weight)

subgraph_nodes = list(G.nodes)
H = G.subgraph(subgraph_nodes)

color_map = {"ai": "red", "gaming": "blue", "rwa": "green", "memes": "purple"}
node_colors = [color_map[G.nodes[n]["clasificacion"]] for n in H.nodes]

plt.figure(figsize=(15, 15))
nx.draw_networkx(H, with_labels=False, node_color=node_colors, node_size=50, edge_color="gray", alpha=0.6)

plt.title("Red de Proyectos Cripto por ClasificaciÃ³n con PonderaciÃ³n")
plt.axis("off")
plt.show()
```
